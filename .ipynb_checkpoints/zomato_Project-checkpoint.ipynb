# Zomato Data Analysis

# Step 1 : Importing data

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

# Step 2 : Creting Dataframe

```python
df=pd.read_csv("Zomato data .csv")
print(df)
```

# Step 3: cleaning Data

### converting rate column to float datatype

```python
def changerate(num):
    value=str(num).split('/')
    value=value[0]
    return float(value)
```

```python
df['rate']=df['rate'].apply(changerate)
print(df.head())
```

```python
df.info()
```

# Step 4 : Analysis

## Type of Resturant

```python
print(df)
```

```python
sns.countplot(x=df['listed_in(type)'],palette='pastel')
plt.xlabel('Types of resturant')
```

### Conclusion- Majority of returant fall in Dining catogory


```python
votes=df.groupby("listed_in(type)")["votes"].sum()
plt.plot(votes,marker='o',c="red")
plt.xlabel(' \n Types of resturant',c="red")
plt.ylabel('Votes',c="red")
plt.show()
```

### Conlusion- Dining Resturant recived most votes

```python
plt.hist(df["rate"],bins=10)
plt.title("Rating Distribution")
plt.xlabel("Rating" ,c="blue")
plt.ylabel("Number of rating",c="blue")
plt.show()
```

### Conclusion- the maximum Rating is between 3.80 to 4.15

```python
sns.countplot(x=df["approx_cost(for two people)"],palette='pastel')
```

 ### Conclusion- Most couple prefer Resturant at price of 300

```python
 sns.boxplot(x="online_order" , y="rate", data=df,palette='pastel')
```

```python
print(df)
```

### Conlusion - offline order recives lower rating as compared to online

```python

pivot_table = df.pivot_table(index='listed_in(type)', columns="online_order", aggfunc="size", fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlOrRd", fmt="d")
plt.xlabel("Online Order")
plt.ylabel("Resturant Type")
plt.show()
```

```python
Conclusion - Dining resturant primarily accept offline order 
```
